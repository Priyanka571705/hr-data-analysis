```python
## Average salary by department

df.groupby("Department")["Salary_INR"].mean().astype(int).sort_values(ascending=False).reset_index()
```

```python
## which Job_title is the highest Avg salary

df.groupby("Job_Title")["Salary_INR"].mean().astype(int).sort_values(ascending=False).reset_index()
```

```python
### How many Employee Resigned & Terminated in each Department

r_emp = df[df["Status"]=="Resigned"].groupby("Department")["Status"].count()
```

```python
t_emp = df[df['Status']=="Terminated"].groupby("Department")["Status"].count()
```

```python
plt.figure(figsize=(10,6))
r_emp.plot(x = r_emp.index,y =r_emp.values,kind="bar",color = "black",label="Resigned")
t_emp.plot(x = t_emp.index,y = t_emp.values, kind="bar",color = "blue", label="Terminated")
plt.legend(title="Employee Status")
plt.ylabel("Employee Count")
plt.show()
```

```python
## what is the avg performance ratings by department ?

pr = df.groupby("Department")["Performance_Rating"].mean().astype(int)
plt.figure(figsize=(8,4))
pr.plot(x=pr.index,y=pr.values,kind="bar",color="green")
plt.title("Average Performance Rating by Department")
plt.ylabel("Average Performance Rating")
plt.show()
```

```python
## How many employees are there in each department

department_counts = df['Department'].value_counts()
plt.figure(figsize=(8,4))
department_counts.plot(kind="bar", color="red")
plt.title("Number of Employees by Department")
plt.ylabel("Number of Employees")
plt.xlabel("Department")
plt.show()
```
